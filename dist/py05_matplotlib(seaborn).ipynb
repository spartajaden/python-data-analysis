```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.countplot(x = 'time', data=tips, alpha=0.7, ax = ax1)
sns.countplot(x = 'day', hue='time', data=tips, alpha=0.7, palette='Set2', ax = ax2)

ax1.set_title("Countplot 시간에 따른 팁 빈도차트")
ax2.set_title("Countplot 날짜와 시간에 따른 팁 빈도차트")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x = 'day', y = 'tip', hue='sex', data=tips, alpha=0.7, ax = ax1)
sns.swarmplot(x = 'day', y = 'tip', hue='sex', data=tips, alpha=0.7, palette='Set2', ax = ax2)

fig.suptitle("Scatter plots 선형회귀의 ON /OFF 비교")
ax1.set_title("선형회귀의 ON")
ax2.set_title("선형회귀의 OFF")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x = 'day', y = 'tip', hue='sex', data=tips, alpha=0.7, ax = ax1)
sns.swarmplot(x = 'day', y = 'tip', hue='sex', data=tips, alpha=0.7, palette='Set2', ax = ax2)
 
ax1.set_title("Strip plot 으로 표현한 날짜별로 분류한 성별과 팁관계")
ax2.set_title("Swarm plot 으로 표현한 날짜별로 분류한 성별과 팁관계")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x ='total_bill', y="tip", color="green", scatter_kws={"s":50 ,"alpha": 0.3}, line_kws={"linestyle":"--","color":"red"} , data= tips , ax = ax1)
sns.regplot(x ='total_bill', y="tip" , data= tips , fit_reg=False,  ax = ax2)

fig.suptitle("Scatter plots 선형회귀의 ON /OFF 비교")
ax1.set_title("선형회귀의 ON ")
ax2.set_title("선형회귀의 OFF")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.boxplot(x = 'day', y="tip", color="green", data=tips, ax = ax1)
sns.barplot(x = 'day', y="tip", data=tips, ax = ax2)

fig.suptitle("Scatter plots 선형회귀의 ON /OFF 비교")
ax1.set_title("선형회귀의 ON")
ax2.set_title("선형회귀의 OFF")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
sns.pairplot(iris , hue="species", markers=["o","s","D"], plot_kws={'alpha': 0.6})
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
fig = plt.figure(figsize=(12,3))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

sns.boxplot(x= "species",y="sepal_length", hue="species", data=iris, ax=ax1)
sns.violinplot(x= "species",y="petal_length",data=iris ,ax=ax2)
sns.histplot(x= "petal_length",data=iris ,hue="species",kde=True, element="step", palette="Set2", ax=ax3)

sns.lmplot(x="sepal_length",y="sepal_width",hue="species",data=iris, markers=["o","s","D"])
plt.title("꽃받침 길이와 넓이의 관계")
sns.lmplot(x="petal_length",y="petal_width",hue="species",data=iris, markers=["o","s","D"])
plt.title("꽃잎 길이와 넓이의 관계")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as pl
exercise = sns.load_dataset('exercise')
exercise.sample(10)
```

```python
plt.figure(figsize=(7,4))
sns.pointplot(x="time",y="pulse",hue="kind",data=exercise, palette="plasma")
plt.title("심박수와 시간별 운동종류와의 관계")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x="time",y="pulse",hue="kind",data=exercise, palette="plasma")
plt.title("심박수와 시간별 운동종류와의 관계")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x="time",y="pulse",hue="kind",data=exercise, palette="Set2")
plt.title("심박수와 시간별 운동종류와의 관계")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x="time",y="pulse",hue="kind",data=exercise, palette="Set1")
plt.title("심박수와 시간별 운동종류와의 관계")
plt.show()
```
