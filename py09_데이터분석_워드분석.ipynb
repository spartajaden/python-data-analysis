```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud  # 🌟 라이브러리를 임포트하여 NameError를 해결합니다.

font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"

# 🌟 WordCloud 생성 시 font_path=font_path 옵션을 꼭 넣어주어야 한글이 깨지지 않습니다!
wd = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate(han_data)

plt.figure(figsize=(12, 6))
plt.imshow(wd, interpolation='bilinear')  # 글자를 선명하게 만들어주는 실무 팁
plt.axis('off')  # 불필요한 테두리 눈금선 숨기기
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from konlpy.tag import Komoran

font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"

# 명사분리하기
komoran = Komoran()
nouns = komoran.nouns(han_data)
nouns_text = " ".join(nouns)

wd = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate(nouns_text)

plt.figure(figsize=(12,6))
plt.imshow(wd, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from konlpy.tag import Komoran
from wordcloud import WordCloud
import matplotlib.pyplot as plt

komoran = Komoran()

# 1. 형태소 분석 결과를 'tokens'라는 이름의 변수로 저장합니다 🌟
tokens = komoran.pos(han_data)  # 형태소 + 품사 태깅

# 2. 저장한 tokens 변수에서 영어(SL)와 명사(NN)를 각각 추출합니다
english = [word for word, pos in tokens if pos == 'SL']
nouns = [word for word, pos in tokens if pos.startswith('NN')]

# 3. 추출한 단어들을 합치고 문자열로 변환
combined = nouns + english
nouns_text = " ".join(combined)

# 4. 워드클라우드 생성
wd = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate(nouns_text)

# 5. 시각화 출력
plt.figure(figsize=(12, 6))

# 🌟 interpolation='bilinear'를 추가하면 글자가 깨지지 않고 부드럽고 선명하게 출력됩니다!
plt.imshow(wd, interpolation='bilinear') 
plt.axis('off') # 불필요한 눈금 테두리 숨기기

plt.show()
```

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
url = "https://fin.land.naver.com/map?center=3AIwyA-2z82v5&zoom=14.260497255506714"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36 Edg/148.0.0.0"}
web = req.get(url, headers=headers)
soup = bs(web.content, 'html.parser')
raw = soup.select(".td.title a")
text =''
for r in raw:
    text += r.text

# print("크롤링 성공! 수집된 글자 수:", len(text))
```

```python
from konlpy.tag import Okt
from konlpy.tag import Komoran
from collections import Counter

okt = Okt()
txt = okt.morphs(text)

# 명사분리하기
# komoran = Komoran()
# txt = komoran.nouns(text)

blacklist = ['광고','\n','특가','오늘','...','ㅋㅋ','....','!!','ㄷㄷ']

filtered_words = [ t for t in txt if len(t) > 1 and t not in blacklist ]
ct = Counter(filtered_words).most_common(10)
ct
```
