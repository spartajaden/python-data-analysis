## 데이터 분석용 함수

```python
import numpy as np
import pandas as pd
data ={
    'korean': [50,60,70,30],
    'math':[30,np.nan,40,20],
}
index = ['홍길순','수퍼걸','쉬헐크','배트걸']
df = pd.DataFrame(data, index=index)
print(df.index)
print(df.columns)
df
```

```python
# count 함수
print(df.count())
print(df.count(axis=1))
```

```python
# max() 함수
print(df.max(axis=0))
print(df.max(axis=1))
```

```python
# min() 함수
print(df.min(axis=0))
print(df.min(axis=1))
```

```python
# sum() 함수
print(df.sum(axis=0))
print(df.sum(axis=1))
```

```python
# mean() 함수
print(df.mean(axis=0))
print(df.mean(axis=1))
print(df.mean(axis=1, skipna=False))
data1 = df.mean(axis=1, skipna=False)
print(data1[~data1.isnull()])
print(data1[data1.notnull()])
```

## 그룹으로 묶기 (groupby)

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'key'    : ['A', 'A', 'A', 'A', 'B', 'C'],
    'data1':[1,2,3,1,2,3],
    'data2':[4,4,6,0,6,1],
    })
print("DataFrame")
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby(['key','data1']).sum().reset_index()
```

```python
df.groupby('key').mean()
```

```python
def filter_mean(x):
    return x['data2'].mean() < 3
df.groupby('key').filter(filter_mean)
# filter_mean(df.groupby('key'))
```

```python
df.groupby('key').aggregate(['min','median','max'])
```

```python
df.groupby('key').aggregate({'data1':'mean','data2':'sum'})
```

```python
# 문제 2 - Only 박철수만 출력하기 
import numpy as np
data0 = [['박철수',60,70]]

data = np.array(data0)
df = pd.DataFrame(data, columns=['이름','국어','수학'])
df
```

```python
# 문제 3 - 수학이 80 이상인 사람 모든 행 출력하기
import numpy as np
import pandas as pd  # pd를 사용하려면 pandas 임포트가 필요합니다.

data0 = [['홍길동', 80, 70, 80], ['박철수', 60, 70, 80], ['김영희', 70, 90, 80]]

# DataFrame을 만들 때 numpy 배열을 거치지 않고 바로 넣는 것이 좋습니다.
df = pd.DataFrame(data0, columns=['이름', '국어', '수학', '영어'])

# 데이터 타입을 숫자로 변환 (수학 컬럼을 비교하기 위함)
df['수학'] = df['수학'].astype(int)

# 수학 점수가 80점 이상인 행 필터링
result = df[df['수학'] >= 80]
result
```

```python
# 문제 4 - 마지막 컬럼에 평균 넣기
import numpy as np
import pandas as pd  # pandas 임포트 확인

data0 = [['홍길동', 80, 70, 80], ['박철수', 60, 70, 80], ['김영희', 70, 90, 80]]

# DataFrame 생성 (숫자 계산을 위해 바로 data0를 넣는 것을 권장합니다)
df = pd.DataFrame(data0, columns=['이름', '국어', '수학', '영어'])

# 1. 국어, 수학, 영어 컬럼을 숫자(int) 타입으로 변환
df[['국어', '수학', '영어']] = df[['국어', '수학', '영어']].astype(int)

# 2. 행(Row) 방향으로 평균을 구해 '평균'이라는 새로운 컬럼에 대입
df['평균'] = df[['국어', '수학', '영어']].mean(axis=1)

df
```
