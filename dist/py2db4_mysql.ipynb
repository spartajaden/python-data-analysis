```python
import mysql.connector

try:
    conn = mysql.connector.connect(
        host="127.0.0.1",
        user="root",
        password="1234",
        database="mydb"
    )
    if conn.is_connected(): print("MySQL 서버에 성공적으로 연결되었습니다!")
except mysql.connector.Error as err:
    print("MySQL 연결 실패:", err)
```

```python
cursor = conn.cursor()
print("SQL 쿼리준비")
```

```python
# 데이터베이스 확인
cursor.execute("SHOW DATABASES")
dbs = cursor.fetchall()
dbs
```

```python
# 테이블 생성
query = """CREATE TABLE IF NOT EXISTS users (
        id INT,
        name VARCHAR(50),
        age INT,
        city VARCHAR(50))
"""

# 자동증가 및 PK, id INT AUTO_INCREMENT PRIMARY KEY
cursor.execute(query)
print("테이블 생성완료")
```

```python
# 테이블 확인
cursor.execute("SHOW TABLES")
tbl = cursor.fetchall()
tbl
```

```python
# 테이블 구조 확인
cursor.execute("DESC users")
desc_tbl = cursor.fetchall()
desc_tbl
```

```python
# 단일 데이터 삽입
sql = f"INSERT INTO users (id, name, age, city) VALUES (%s,%s,%s,%s)"
values = (1, "Alice", 22, "Seoul")
cursor.execute(sql, values)
conn.commit()
print("1개 데이터 삽입완료", cursor.lastrowid)
```

```python
# 복수 데이터 삽입
sql = f"INSERT INTO users (id, name, age, city) VALUES (%s,%s,%s,%s)"
values = (1, "Alice", 22, "Seoul")
cursor.execute(sql, values)
conn.commit()
print("1개 데이터 삽입완료", cursor.lastrowid)
```

```python
# 단일 데이터 삽입
sql = f"INSERT INTO users (id, name, age, city) VALUES (%s,%s,%s,%s)"

values = (1, "Alice", 22, "Seoul")
m_values = [
    (2, "Bob", 30, "Busan"),
    (2, "Charlie", 28, "Incheon"),
    (2, "Diana", 35, "Daegu"),
    (2, "Ethan", 26, "Gwangju"),
    (2, "Fiona", 40, "Jeju"),
    (2, "George", 33, "Seoul"),
    (2, "Hana", 27, "Suwon"),
    (2, "Ian", 29, "Ulsan"),
    (2, "Jisoo", 24, "Daejeon"),
    (2, "Kevin", 31, "Changwon"),
]
cursor.execute(sql, values)
print("1개 데이터 삽입완료", cursor.lastrowid)

cursor.executemany(sql, m_values)
print("여러개 데이터 삽입완료.")
conn.commit()
```

```python
# 데이터 조회
sql = "SELECT * FROM users"
filters = " WHERE age >= 30"
cursor.execute(sql + filters)
# 방법 1:
# for n in cursor:
#     print(n)
# 방법 2:
data1 = cursor.fetchall()
for d in data1:
    print(d)
```

```python
# 업데이트
sql = "UPDATE users SET age = %s WHERE name = %s"
values = (31, "Bob")
cursor.execute(sql, values)
conn.commit()
print("수정된 행 개수:", cursor.rowcount)
```
