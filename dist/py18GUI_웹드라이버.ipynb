```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By          
from selenium.webdriver.common.keys import Keys        
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup as bs
import pandas as pd
import time as tt

# 1. 크롬 옵션 설정
options = Options()
options.add_argument("--window-size=1920,1080") 
# options.add_argument("--no-sandbox")
# options.add_argument("--disable-dev-shm-usage")
# options.add_argument("--remote-allow-origins=*")

# 💡 Selenium 4.6+ 의 Selenium Manager가 Chrome 버전에 맞는 드라이버를 자동으로 찾아 다운로드합니다.
driver = webdriver.Chrome(options=options)

# 클릭을 편리하게 해주는 gogo 함수 정의
def gogo(sel):
    driver.find_element(By.CSS_SELECTOR, sel).click()

# 3. 사이트 접속
driver.get('https://tenpercentcoffee.com/')
print("페이지 제목:", driver.title)
tt.sleep(2) 

# 4. 마우스 호버할 요소 찾기 및 실행
hover_element = driver.find_element(By.CSS_SELECTOR, "#header > div")
action = ActionChains(driver)
action.move_to_element(hover_element).perform()
print('메뉴 여는중...')
tt.sleep(1)

# 5. 매장 찾기 메뉴 클릭
gogo("#header > div > div > ul > li:nth-child(4) > ul > li > a")
print('매장 위치 검색 페이지 이동 중...')
tt.sleep(3) 

print("매장 위치 로드 완료")

# 6. 검색창 클릭 후 '부산광역시' 입력 및 엔터
gogo("#q")
tt.sleep(1)
driver.find_element(By.CSS_SELECTOR, "#q").send_keys("서울특별시")
driver.find_element(By.CSS_SELECTOR, "#q").send_keys(Keys.ENTER) 
print('서울특별시 검색어 입력 완료')

# 💡 비동기 데이터가 로드될 때까지 최대 10초간 대기
try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#area_list > li"))
    )
    print("매장 검색 결과 로딩 완료!")
except:
    print("로딩 시간이 초과되었거나 검색 결과가 없습니다.")

# 7. BeautifulSoup으로 데이터 파싱
html = driver.page_source
soup = bs(html, 'html.parser')

# 매장 리스트 요소 선택
contents = soup.select('#area_list > li')

temp = []
print(f"찾은 매장 개수: {len(contents)}개")

for c in contents:
    title_element = c.select_one('strong') 
    
    if title_element:
        store_name = title_element.text.strip()
        print(f'성공! 매장명: {store_name}')
        temp.append(store_name)
    else:
        store_info = c.text.strip().replace('\n', ' ')
        print(f'성공! 정보: {store_info}')
        temp.append(store_info)

# 결과 출력
print("\n--- 수집된 매장 리스트 ---")
print(temp)

# 8. 브라우저 종료
# driver.quit()
```

```python
from datetime import datetime
from hdfs import InsecureClient

timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
csv_file = f"tencafe{timestamp}.csv"
df.to_csv(csv_file, index=True)
print("csv 변환완료")

hdfs = InsecureClient('http://namenode:9870',user='hadoop')
hdfs_path = f"/dataset/{csv_file}"
try:
    hdfs.upload(hdfs_path, csv_file)
    print("하둡에 업로드 완료!")
except Exception as e:
    print("업로드 실패!", e)
```
